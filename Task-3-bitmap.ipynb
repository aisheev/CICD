```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def display_image(image_data, title):
    plt.imshow(image_data)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load the image
image_path = '/Users/ash/Downloads/buttercup.jpg'  
original_image = Image.open(image_path)
image_data = np.array(original_image)

# Display the original image
display_image(image_data, 'Original Image')

# Encryption key
key = 100 

# Encrypt the image (simple XOR operation)
encrypted_image_data = image_data ^ key

# Display the encrypted image
display_image(encrypted_image_data, 'Encrypted Image')

# Decrypt the image (XOR operation again)
decrypted_image_data = encrypted_image_data ^ key

# Display the decrypted image
display_image(decrypted_image_data, 'Decrypted Image')
```
