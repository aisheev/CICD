```python
#Encyrpting Alice Malayalam Message

def cipher_malayalam(text, shift):
    result = ""
    
    
    start_unicode = ord('\u0D05') # Define the start and end of the Malayalam unicode range - \u0D05 represents alphabet അ
    end_unicode = ord('\u0D39') #represents ഹ
    num_letters = end_unicode - start_unicode + 1 #setting the range 
    
    for char in text:
        if start_unicode <= ord(char) <= end_unicode: 
            shifted = start_unicode + ((ord(char) - start_unicode + shift) % num_letters) # Calculate the shifted code with modulo arithmetic
            result += chr(shifted)
        else:
            
            result += char # If it's not a Malayalam letter, just append the character as is

    return result


alice_mallu_message = """തെളിമാനം മഴവില്ലിൻ നിറമണിയും നേരം
നിറമാർന്നൊരു കനവെന്നിൽ തെളിയുന്ന പോലെ
പുഴയോരം തഴുകുന്നീ തണുവീറൻ കാറ്റും
പുളകങ്ങൾ ഇഴനെയ്തൊരു കുഴലൂതിയ പോലെ
കുളിരേകും കനവെന്നിൽ കതിരാടിയ കാലം
മനതാരിൽ മധുമാസം തളിരാടിയ നേരം
അകമരുവും മയിലിണകൾ തുയിലുണരും കാലം
എൻ അകതാരിൽ അനുരാഗം പകരുന്ന യാമം

അഴകേ... അഴകിൽ തീർത്തൊരു ശിലയഴകേ
മലരേ... എന്നുയിരിൽ വിടരും പനിമലരേ...

മലരേ നിന്നെ കാണാതിരുന്നാൽ
മിഴിവേകിയ നിറമെല്ലാം മായുന്നപോലെ
അലിവോടെന്നരികത്തിന്നണയാതിരുന്നാൽ
അഴകേകിയ കനവെല്ലാം അകലുന്നപോലെ

ഞാനെന്റെ ആത്മാവിനാഴത്തിനുള്ളിൽ
അതിലോലമാരോരുമറിയാതെ സൂക്ഷിച്ച
താളങ്ങൾ രാഗങ്ങൾ ഈണങ്ങളായി
ഓരോരോ വർണ്ണങ്ങളായ്

ഇടറുന്നൊരെന്റെ ഇടനെഞ്ചിനുള്ളിൽ
പ്രണയത്തിൻ മഴയായ് നീ പൊഴിയുന്നീ നാളിൽ
തളരുന്നൊരെന്റെ തനുതോറും നിന്റെ
അലതല്ലും പ്രണയത്താലുണരും മലരേ...

അഴകേ...


കുളിരേകും കനവെന്നിൽ കതിരാടിയ കാലം
മനതാരിൽ മധുമാസം തളിരാടിയ നേരം
അകമരുവും മയിലിണകൾ തുയിലുണരും കാലം
എൻ അകതാരിൽ അനുരാഗം പകരുന്ന യാമം


അഴകേ... അഴകിൽ തീർത്തൊരു ശിലയഴകേ
മലരേ... എന്നുയിരിൽ വിടരും പനിമലരേ..."""  # Malayalam text to be encrypted

shift_amount = 7  # The shift (key) for the cipher

encrypted_alice = cipher_malayalam(alice_mallu_message, shift_amount)
print(f"Encrypted text: {encrypted_alice}")
print()

```
